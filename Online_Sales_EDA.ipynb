```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Diwali.txt')
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.drop(['Status', 'unnamed1'], axis = 1, inplace =True)
```

```python
df.info()
```

```python
df.dropna(inplace = True)
```

```python
df.info()
```

```python
df['Amount'].dtype
```

```python
#Convert - float data type into Integer 
df['Amount'] = df['Amount'].astype('int')
```

```python
df.info()
```

```python
df.head()
```

```python
df_age = df.groupby('Age Group').Amount.sum()
df_age = df_age.reset_index()
df_age
```

```python
df.rename(columns = {'Marital_Status' : 'Shaadi'}, inplace = True)
cplot = sns.countplot(x = 'Shaadi', data = df, hue = 'Gender')
plt.xlabel('Marrige Status')
plt.ylabel('Toatl Count of Customers')
plt.title('Married vs Unmarried Customers')
plt.bar_label(cplot.containers[0])
plt.bar_label(cplot.containers[1])
```

Total of 4573 Unmarried Females are major Customers in comparison of both male and Female.
on the Other Side, 1462 Married Male are counted much less as compared to others.

```python
df.head()
```

```python
sns.countplot(x = 'Age Group', data = df,  hue = 'Gender')
```

```python
g = df.groupby(['Age Group', 'Gender']).Orders.count()
g = g.reset_index()
gplt = sns.barplot(x = 'Age Group', y = 'Orders', data = g, hue = 'Gender')
plt.bar_label(gplt.containers[0])
plt.bar_label(gplt.containers[1])
```

```python
# Female in Age Group 26-35 are likely to order more compared to others.
```

```python
#Total Amount vs Age Group

total_sales = df.groupby('Age Group', as_index = False).Amount.sum().sort_values(by = 'Amount' , ascending = False)
plt.pie(x = 'Amount', data = total_sales, labels = 'Age Group', autopct = '%1.1f%%')
plt.legend(loc = 'upper right')
```

```python
# in total, Age Group between 26-35 are more likely to buy from the websites as their Total Sales amount accounted around 40%.
```

```python
df.head()
```

```python
gocc = df.groupby(['Occupation','Gender'], as_index= False).Amount.sum().sort_values(by = 'Amount', ascending = False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'Occupation', y = 'Amount', data = gocc , ci = None )
```

```python
gocc = df.groupby(['Occupation','Gender'], as_index= False).Amount.sum().sort_values(by = 'Amount', ascending = True)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'Occupation', y = 'Amount', data = gocc , ci = None )
```

```python
# Customers are likely to order online they are from It, Avialtion and Helathcare Sector
# The Customers are less likely to order online they are from Agriculture, Construction and textile.
```

```python
df.head()
```

```python
gstate = df.groupby('State', as_index = False).Orders.count().sort_values(by = 'Orders', ascending = False)
gstateplot = sns.barplot(x ='State', y = 'Orders', data = gstate)
plt.xticks(rotation = 45)
plt.bar_label(gstateplot.containers[0])
```

```python
gprocat = df.groupby('Product_Category', as_index = False).Orders.count().sort_values(by = 'Orders', ascending = False)
gproplt = sns.barplot(x ='Product_Category', y = 'Orders', data = gprocat)
plt.xticks(rotation = 45)
plt.bar_label(gproplt.containers[0])
```

```python
# by seeing above this two graphs, we can see that the Product which falls into Clothing & Apparel , Food and Electronics and Gadgets are more likely to be purchased through online.
#Top 5 states accounted for big orders number are as follws : Uttar Pradesh, Maarashtra and Karnatak
```

```python
gzone = df.groupby('Zone', as_index = False).Amount.sum().sort_values(by = 'Amount', ascending = False)
exp = [0.7, 0.3, 0.3, 0.3, 0.3]
plt.pie(x = 'Amount', labels = 'Zone', data = gzone, autopct = '%1.1f%%', explode = exp)
plt.legend(loc = 'center')
```

```python
f1 = {'family' : 'Arial', 'size' : 20}

gzprdt = df.groupby('Zone', as_index = False).Product_Category.count().sort_values(by= 'Product_Category', ascending = False)
gzprdtplt = sns.barplot(x = 'Zone', y = 'Product_Category', data = gzprdt, color = 'c')
plt.title('Total_Products Vs Zone', fontdict = f1)
plt.xlabel('Zones', fontdict= f1)
plt.ylabel('Total Product Counts by Category' , fontdict = f1)
plt.bar_label(gzprdtplt.containers[0],label_type='center', fontsize = 20, color = 'r')
```

```python
gz = df.groupby(['Zone','State']).Orders.sum()
gz = gz.reset_index()
sns.barplot(x = 'Zone', y = 'Orders', data  = gz, ci = None)
```

```python
# The Zonewise Orders shows that the Central and Western Zone has more influence on online ordering.
#on other side, moderate to less influence show easter and northern zones.
```
